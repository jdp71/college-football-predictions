```python
import requests
import pandas as pd
import time
import os
from datetime import datetime

API_KEY = "YOUR_CFBD_API_KEY"  # Replace with your actual key
HEADERS = {"Authorization": f"Bearer K4cogGMiQm5XURMHztnZpfuE409un7xxKEkoACOhlg2Rn+y94ahRMOxXZ3d+92V6"}
BASE_URL = "https://api.collegefootballdata.com"
SEASONS = [2020, 2021, 2022, 2023, 2024]  # Add/remove years as needed
OUTDIR = "cfbd_data"
os.makedirs(OUTDIR, exist_ok=True)

def fetch(endpoint, params, filename, description="", year=None):
    """
    Fetch data from CFBD API with better error handling
    """
    # Add year to filename if provided
    if year and year != "":
        name, ext = os.path.splitext(filename)
        filename = f"{name}_{year}{ext}"
    url = f"{BASE_URL}/{endpoint}"
    print(f"\n{'='*50}")
    print(f"Fetching: {description or endpoint}")
    print(f"URL: {url}")
    print(f"Params: {params}")
    
    try:
        resp = requests.get(url, headers=HEADERS, params=params)
        print(f"Response status: {resp.status_code}")
        
        if resp.status_code == 400:
            print(f"Bad Request Error. Response text: {resp.text}")
            return None
        elif resp.status_code == 401:
            print(f"Unauthorized. Check your API key.")
            return None
        elif resp.status_code == 429:
            print(f"Rate limit exceeded. Waiting longer...")
            time.sleep(5)
            return fetch(endpoint, params, filename, description)  # Retry
        
        resp.raise_for_status()
        data = resp.json()
        
        if not data:
            print(f"No data returned for {endpoint}")
            return None
            
        df = pd.json_normalize(data)
        
        filepath = f"{OUTDIR}/{filename}"
        df.to_csv(filepath, index=False)
        print(f"✓ Saved: {filename} ({len(df)} rows, {len(df.columns)} columns)")
        
        # Show sample of data
        if len(df) > 0:
            print(f"Sample columns: {list(df.columns)[:5]}")
        
        time.sleep(1)  # Rate limiting
        return df
        
    except requests.exceptions.HTTPError as e:
        print(f"❌ HTTP Error for {endpoint}: {e}")
        if 'resp' in locals():
            print(f"Response content: {resp.text}")
        return None
    except Exception as e:
        print(f"❌ Error fetching {endpoint}: {e}")
        return None

def fetch_all_endpoints():
    """
    Fetch data from all specified CFBD endpoints for multiple years
    """
    print(f"Starting CFBD data collection for seasons: {SEASONS}")
    print(f"Output directory: {OUTDIR}")
    print(f"Timestamp: {datetime.now()}")
    
    total_successful_fetches = 0
    total_endpoints_attempted = 0
    
    # Process each season
    for season in SEASONS:
        print(f"\n{'*'*80}")
        print(f"PROCESSING SEASON: {season}")
        print(f"{'*'*80}")
        
        successful_fetches = 0
        total_endpoints = 0
    
        # Basic team and game data
        print(f"\n{'#'*60}")
        print("# BASIC DATA ENDPOINTS")
        print(f"{'#'*60}")
        
        endpoints_basic = [
            ("teams", {}, "teams.csv", "All Teams", ""),  # No year needed for teams
            ("teams/fbs", {}, "teams_fbs.csv", "FBS Teams Only", ""),  # No year needed for FBS teams
            ("games", {"year": season}, "games.csv", f"Games for {season}", season),
            ("drives", {"year": season}, "drives.csv", f"Drives for {season}", season),
        ]
        
        for endpoint, params, filename, description, year in endpoints_basic:
            total_endpoints += 1
            result = fetch(endpoint, params, filename, description, year)
            if result is not None:
                successful_fetches += 1
    
        # Statistics endpoints
        print(f"\n{'#'*60}")
        print("# STATISTICS ENDPOINTS")
        print(f"{'#'*60}")
        
        endpoints_stats = [
            ("stats/season", {"year": season}, "team_stats.csv", f"Team Season Stats {season}", season),
            ("stats/player/season", {"year": season}, "player_stats.csv", f"Player Season Stats {season}", season),
            ("stats/season/advanced", {"year": season}, "advanced_stats.csv", f"Advanced Team Stats {season}", season),
        ]
        
        for endpoint, params, filename, description, year in endpoints_stats:
            total_endpoints += 1
            result = fetch(endpoint, params, filename, description, year)
            if result is not None:
                successful_fetches += 1
    
        # Ratings endpoints
        print(f"\n{'#'*60}")
        print("# RATINGS ENDPOINTS")
        print(f"{'#'*60}")
        
        endpoints_ratings = [
            ("ratings/sp", {"year": season}, "ratings_sp.csv", f"S&P+ Ratings {season}", season),
            ("ratings/sp/conferences", {"year": season}, "ratings_sp_conferences.csv", f"S&P+ Conference Ratings {season}", season),
            ("ratings/srs", {"year": season}, "ratings_srs.csv", f"SRS Ratings {season}", season),
            ("ratings/elo", {"year": season}, "ratings_elo.csv", f"Elo Ratings {season}", season),
            ("ratings/fpi", {"year": season}, "ratings_fpi.csv", f"FPI Ratings {season}", season),
        ]
        
        for endpoint, params, filename, description, year in endpoints_ratings:
            total_endpoints += 1
            result = fetch(endpoint, params, filename, description, year)
            if result is not None:
                successful_fetches += 1
    
        # Plays-related endpoints (these might need special handling)
        print(f"\n{'#'*60}")
        print("# PLAYS METADATA ENDPOINTS")
        print(f"{'#'*60}")
        
        endpoints_plays_meta = [
            ("plays/types", {}, "plays_types.csv", "Play Types Reference", ""),  # No year needed for types
            ("plays/stats", {"year": season}, "plays_stats.csv", f"Play Stats {season}", season),
            ("plays/stats/types", {}, "plays_stats_types.csv", "Play Stats Types Reference", ""),  # No year needed for types
        ]
        
        for endpoint, params, filename, description, year in endpoints_plays_meta:
            total_endpoints += 1
            result = fetch(endpoint, params, filename, description, year)
            if result is not None:
                successful_fetches += 1
    
        # Team matchup data
        print(f"\n{'#'*60}")
        print("# TEAM MATCHUP ENDPOINTS")
        print(f"{'#'*60}")
        
        # Get FBS teams first for matchup data
        teams_df = fetch("teams/fbs", {}, "teams_fbs_for_matchup.csv", "FBS Teams for Matchup", "")
        if teams_df is not None and not teams_df.empty:
            total_endpoints += 1
            successful_fetches += 1
            
            # Try to get matchup data for a few teams
            if 'school' in teams_df.columns:
                sample_teams = teams_df['school'].head(3).tolist()
                print(f"Sample teams for matchup: {sample_teams}")
                
                for i, team in enumerate(sample_teams):
                    total_endpoints += 1
                    result = fetch("teams/matchup", 
                                 {"team1": team, "team2": sample_teams[(i+1) % len(sample_teams)]}, 
                                 f"matchup_{team.replace(' ', '_')}.csv", 
                                 f"Matchup: {team}", 
                                 season)
                    if result is not None:
                        successful_fetches += 1
    
        # Game-specific plays data (sample)
        print(f"\n{'#'*60}")
        print("# GAME PLAYS DATA (SAMPLE)")
        print(f"{'#'*60}")
        
        # Try to get plays for a few games
        try:
            games_file = f"{OUTDIR}/games_{season}.csv"
            if os.path.exists(games_file):
                games_df = pd.read_csv(games_file)
                if not games_df.empty and 'id' in games_df.columns:
                    sample_games = games_df['id'].head(3)  # Just try 3 games
                    print(f"Attempting plays data for {len(sample_games)} sample games from {season}...")
                    
                    for game_id in sample_games:
                        total_endpoints += 1
                        result = fetch("plays", 
                                     {"gameId": int(game_id)}, 
                                     f"plays_game_{game_id}.csv", 
                                     f"Plays for Game {game_id}", 
                                     season)
                        if result is not None:
                            successful_fetches += 1
                else:
                    print(f"No games data available for {season} plays extraction")
            else:
                print(f"Games file not found for {season} - skipping individual game plays")
        except Exception as e:
            print(f"Error processing games file for {season}: {e}")
        
        # Recruiting data
        print(f"\n{'#'*60}")
        print("# RECRUITING DATA")
        print(f"{'#'*60}")
        
        total_endpoints += 1
        recruiting_result = fetch("recruiting/players", {"year": season}, "recruiting.csv", f"Recruiting {season}", season)
        if recruiting_result is not None:
            successful_fetches += 1
        
        # Season summary
        print(f"\n{'-'*50}")
        print(f"SEASON {season} SUMMARY")
        print(f"{'-'*50}")
        print(f"Endpoints attempted: {total_endpoints}")
        print(f"Successful fetches: {successful_fetches}")
        print(f"Success rate: {(successful_fetches/total_endpoints)*100:.1f}%")
        
        total_endpoints_attempted += total_endpoints
        total_successful_fetches += successful_fetches
    
    # Live plays (only fetch once, not per season)
    print(f"\n{'#'*60}")
    print("# LIVE DATA ENDPOINTS")
    print(f"{'#'*60}")
    
    total_endpoints_attempted += 1
    live_result = fetch("live/plays", {}, "live_plays.csv", "Live Plays (if available)", "")
    if live_result is not None:
        total_successful_fetches += 1
    
    # Final Summary
    print(f"\n{'='*80}")
    print("FINAL COLLECTION SUMMARY")
    print(f"{'='*80}")
    print(f"Seasons processed: {SEASONS}")
    print(f"Total endpoints attempted: {total_endpoints_attempted}")
    print(f"Total successful fetches: {total_successful_fetches}")
    print(f"Overall success rate: {(total_successful_fetches/total_endpoints_attempted)*100:.1f}%")
    print(f"Output directory: {OUTDIR}")
    print(f"Completed at: {datetime.now()}")
    
    # List all created files
    try:
        files = [f for f in os.listdir(OUTDIR) if f.endswith('.csv')]
        print(f"\nCreated files ({len(files)}):")
        for file in sorted(files):
            filepath = os.path.join(OUTDIR, file)
            size = os.path.getsize(filepath)
            print(f"  - {file} ({size:,} bytes)")
    except Exception as e:
        print(f"Error listing files: {e}")

if __name__ == "__main__":
    fetch_all_endpoints()
```

```python
import requests
import pandas as pd
import json
from datetime import datetime
import time

# College Football Data API approach
def create_2025_schedule_from_cfbd():
    """
    Create 2025 schedule CSV using College Football Data API
    """
    API_KEY = "YOUR_CFBD_API_KEY"  # Replace with your actual key
    HEADERS = {"Authorization": f"Bearer K4cogGMiQm5XURMHztnZpfuE409un7xxKEkoACOhlg2Rn+y94ahRMOxXZ3d+92V6"}
    BASE_URL = "https://api.collegefootballdata.com"
    
    print("Fetching 2025 college football schedule from CFBD API...")
    
    try:
        # Get 2025 games
        url = f"{BASE_URL}/games"
        params = {"year": 2025}
        
        response = requests.get(url, headers=HEADERS, params=params)
        response.raise_for_status()
        
        games_data = response.json()
        
        if not games_data:
            print("No 2025 games found in CFBD API yet.")
            return create_sample_2025_schedule()
        
        # Convert to desired format
        schedule_data = []
        
        for game in games_data:
            # Parse date
            if game.get('start_date'):
                try:
                    date_obj = datetime.fromisoformat(game['start_date'].replace('Z', '+00:00'))
                    formatted_date = date_obj.strftime('%Y-%m-%d')
                except:
                    formatted_date = game.get('start_date', 'TBD')
            else:
                formatted_date = 'TBD'
            
            schedule_data.append({
                'Date': formatted_date,
                'Home Team': game.get('home_team', ''),
                'Visiting Team': game.get('away_team', ''),
                'Week': game.get('week', ''),
                'Season Type': game.get('season_type', 'regular'),
                'Neutral Site': 'Yes' if game.get('neutral_site') else 'No',
                'Conference Game': 'Yes' if game.get('conference_game') else 'No'
            })
        
        df = pd.DataFrame(schedule_data)
        df = df.sort_values(['Date', 'Home Team'])
        
        # Save to CSV
        filename = '2025_ncaa_football_schedule.csv'
        df.to_csv(filename, index=False)
        
        print(f"✓ Created {filename} with {len(df)} games")
        print(f"Date range: {df['Date'].min()} to {df['Date'].max()}")
        print(f"Sample games:")
        print(df[['Date', 'Home Team', 'Visiting Team']].head())
        
        return df
        
    except requests.exceptions.RequestException as e:
        print(f"Error fetching from CFBD API: {e}")
        print("Creating sample schedule instead...")
        return create_sample_2025_schedule()

def create_sample_2025_schedule():
    """
    Create a sample 2025 schedule with realistic matchups
    """
    print("Creating sample 2025 NCAA football schedule...")
    
    # Major teams and their typical schedules
    teams_by_conference = {
        'SEC': ['Alabama', 'Georgia', 'Florida', 'LSU', 'Auburn', 'Tennessee', 'South Carolina', 'Kentucky', 
                'Mississippi', 'Mississippi State', 'Arkansas', 'Vanderbilt', 'Missouri', 'Texas A&M', 'Texas', 'Oklahoma'],
        'Big Ten': ['Ohio State', 'Michigan', 'Penn State', 'Wisconsin', 'Iowa', 'Michigan State', 'Minnesota', 
                   'Nebraska', 'Northwestern', 'Illinois', 'Indiana', 'Purdue', 'Maryland', 'Rutgers', 'USC', 'UCLA', 'Oregon', 'Washington'],
        'ACC': ['Clemson', 'Florida State', 'Miami', 'North Carolina', 'NC State', 'Virginia Tech', 'Virginia', 
                'Duke', 'Wake Forest', 'Georgia Tech', 'Pittsburgh', 'Syracuse', 'Boston College', 'Louisville'],
        'Big 12': ['Kansas State', 'Kansas', 'Oklahoma State', 'Texas Tech', 'Baylor', 'TCU', 'West Virginia', 
                  'Cincinnati', 'Houston', 'UCF', 'Iowa State', 'BYU', 'Utah', 'Colorado', 'Arizona', 'Arizona State'],
        'Pac-12': ['Stanford', 'California', 'Oregon State', 'Washington State'],
        'Group of 5': ['Boise State', 'Fresno State', 'San Diego State', 'UNLV', 'Air Force', 'Colorado State',
                      'Memphis', 'SMU', 'Tulane', 'Navy', 'Army', 'Notre Dame']
    }
    
    schedule_data = []
    
    # Week 0 (August 23, 2025)
    week_0_games = [
        ('2025-08-23', 'Dublin, Ireland', 'Kansas State', 'Iowa State'),  # Aer Lingus Classic
        ('2025-08-23', 'Nicholls', 'UIW'),
        ('2025-08-23', 'Kansas', 'Fresno State'),
        ('2025-08-23', 'Western Kentucky', 'Sam Houston'),
        ('2025-08-23', 'Hawaii', 'Stanford'),
    ]
    
    for date, home, away in week_0_games:
        schedule_data.append({
            'Date': date,
            'Home Team': home,
            'Visiting Team': away,
            'Week': 0,
            'Season Type': 'regular',
            'Neutral Site': 'Yes' if 'Ireland' in home else 'No',
            'Conference Game': 'No'
        })
    
    # Week 1 (August 30, 2025) - Major kickoff games
    week_1_games = [
        ('2025-08-30', 'Ohio State', 'Texas'),  # Big matchup
        ('2025-08-30', 'Florida State', 'Alabama'),  # Chick-fil-A Kickoff
        ('2025-08-30', 'Colorado', 'Georgia Tech'),
        ('2025-08-30', 'Baylor', 'Auburn'),
        ('2025-08-30', 'Tennessee', 'Syracuse'),  # Aflac Kickoff
        ('2025-08-30', 'Oregon', 'Montana State'),
        ('2025-08-30', 'Michigan', 'Bowling Green'),
        ('2025-08-30', 'Georgia', 'UT Martin'),
        ('2025-08-30', 'Penn State', 'West Virginia'),
        ('2025-08-30', 'USC', 'Hawaii'),
    ]
    
    for date, home, away in week_1_games:
        schedule_data.append({
            'Date': date,
            'Home Team': home,
            'Visiting Team': away,
            'Week': 1,
            'Season Type': 'regular',
            'Neutral Site': 'No',
            'Conference Game': 'No'
        })
    
    # Generate more weeks with realistic matchups
    import random
    random.seed(42)  # For reproducible results
    
    # Weeks 2-12 (regular season)
    for week in range(2, 13):
        # Calculate date (roughly every Saturday)
        base_date = datetime(2025, 8, 30)
        week_date = base_date.replace(day=base_date.day + (week-1)*7)
        date_str = week_date.strftime('%Y-%m-%d')
        
        # Generate conference games and rivalries
        games_this_week = []
        
        if week == 2:  # Big non-conference games
            games_this_week = [
                (date_str, 'Alabama', 'Wisconsin'),
                (date_str, 'Notre Dame', 'Texas A&M'),
                (date_str, 'Oklahoma', 'Michigan'),
                (date_str, 'Oregon', 'Oklahoma State'),
                (date_str, 'Georgia', 'Clemson'),
                (date_str, 'Florida', 'Miami'),
            ]
        
        elif week == 12:  # Rivalry week
            games_this_week = [
                (date_str, 'Michigan', 'Ohio State'),  # The Game
                (date_str, 'Alabama', 'Auburn'),  # Iron Bowl
                (date_str, 'Georgia', 'Georgia Tech'),  # Clean, Old-Fashioned Hate
                (date_str, 'Florida', 'Florida State'),  # Sunshine Showdown
                (date_str, 'Texas', 'Texas A&M'),  # Lone Star Showdown
                (date_str, 'USC', 'UCLA'),  # Battle for Los Angeles
            ]
        
        else:  # Regular conference games
            # Sample some conference matchups
            sec_matchups = [
                (date_str, 'LSU', 'Alabama'),
                (date_str, 'Georgia', 'Florida'),
                (date_str, 'Tennessee', 'Kentucky'),
                (date_str, 'Auburn', 'Mississippi State'),
            ]
            
            big_ten_matchups = [
                (date_str, 'Penn State', 'Michigan State'),
                (date_str, 'Wisconsin', 'Iowa'),
                (date_str, 'Nebraska', 'Northwestern'),
                (date_str, 'Oregon', 'Washington'),
            ]
            
            games_this_week = sec_matchups[:2] + big_ten_matchups[:2]
        
        for date, home, away in games_this_week:
            # Determine if it's a conference game
            conf_game = False
            for conf, teams in teams_by_conference.items():
                if home in teams and away in teams:
                    conf_game = True
                    break
            
            schedule_data.append({
                'Date': date,
                'Home Team': home,
                'Visiting Team': away,
                'Week': week,
                'Season Type': 'regular',
                'Neutral Site': 'No',
                'Conference Game': 'Yes' if conf_game else 'No'
            })
    
    # Add some bowl games and playoffs (December 2025 - January 2026)
    bowl_games = [
        ('2025-12-20', 'TBD', 'TBD'),  # CFP First Round
        ('2025-12-21', 'TBD', 'TBD'),  # CFP First Round
        ('2025-12-21', 'TBD', 'TBD'),  # CFP First Round
        ('2025-12-21', 'TBD', 'TBD'),  # CFP First Round
        ('2025-12-31', 'TBD', 'TBD'),  # Cotton Bowl
        ('2026-01-01', 'TBD', 'TBD'),  # Rose Bowl
        ('2026-01-01', 'TBD', 'TBD'),  # Sugar Bowl
        ('2026-01-01', 'TBD', 'TBD'),  # Orange Bowl
        ('2026-01-19', 'TBD', 'TBD'),  # CFP National Championship
    ]
    
    for date, home, away in bowl_games:
        schedule_data.append({
            'Date': date,
            'Home Team': home,
            'Visiting Team': away,
            'Week': 'Bowl/Playoff',
            'Season Type': 'postseason',
            'Neutral Site': 'Yes',
            'Conference Game': 'No'
        })
    
    # Create DataFrame
    df = pd.DataFrame(schedule_data)
    df = df.sort_values(['Date', 'Home Team'])
    
    # Save to CSV with requested columns
    output_df = df[['Date', 'Home Team', 'Visiting Team']].copy()
    filename = '2025_ncaa_football_schedule.csv'
    output_df.to_csv(filename, index=False)
    
    print(f"✓ Created {filename} with {len(output_df)} games")
    print(f"Date range: {output_df['Date'].min()} to {output_df['Date'].max()}")
    print(f"\nSample games:")
    print(output_df.head(10).to_string(index=False))
    
    return output_df

def scrape_espn_schedule():
    """
    Alternative: Scrape ESPN's 2025 schedule
    Note: This is more complex and may require handling JavaScript
    """
    print("Attempting to scrape ESPN schedule...")
    
    try:
        import requests
        from bs4 import BeautifulSoup
        
        # This would require more sophisticated scraping
        # ESPN loads content dynamically, so you'd need selenium or similar
        print("ESPN scraping requires advanced tools like Selenium.")
        print("Using sample data instead.")
        
        return create_sample_2025_schedule()
        
    except ImportError:
        print("BeautifulSoup not available. Using sample data.")
        return create_sample_2025_schedule()

# Main execution
if __name__ == "__main__":
    print("Creating 2025 NCAA Football Schedule CSV")
    print("=" * 50)
    
    # Try CFBD API first, then fall back to sample data
    try:
        schedule_df = create_2025_schedule_from_cfbd()
    except Exception as e:
        print(f"Error with CFBD approach: {e}")
        schedule_df = create_sample_2025_schedule()
    
    print(f"\n✓ Schedule CSV created successfully!")
    print(f"  Columns: Date, Home Team, Visiting Team")
    print(f"  Total games: {len(schedule_df)}")
    print(f"  File: 2025_ncaa_football_schedule.csv")
```

```python
# Cell 3: Data Loading Function
def load_cfb_data(years, data_dir):
    """Load and combine college football data from multiple years"""
    data = {}
    
    for data_type in ['games', 'team_stats', 'advanced_stats', 'ratings_sp', 'ratings_elo', 'ratings_fpi']:
        combined_data = []
        
        for year in years:
            try:
                if data_type in ['teams', 'teams_fbs']:
                    # These don't have year suffix
                    df = pd.read_csv(f'{data_dir}/{data_type}.csv')
                else:
                    df = pd.read_csv(f'{data_dir}/{data_type}_{year}.csv')
                df['year'] = year
                combined_data.append(df)
                print(f"✓ Loaded {data_type} for {year}: {len(df)} rows")
            except FileNotFoundError:
                print(f"✗ {data_type}_{year}.csv not found")
        
        if combined_data:
            data[data_type] = pd.concat(combined_data, ignore_index=True)
        else:
            data[data_type] = pd.DataFrame()
    
    return data
```

```python
# Cell 6: Feature Engineering
# Cell 6: Feature Engineering (Updated for Games Without Scores)
def create_team_features(team_name, year, cfb_data):
    """Create comprehensive features for a team in a given year"""
    features = {}
    
    # Basic team stats - handle different data structures
    team_stats = cfb_data['team_stats']
    if not team_stats.empty:
        team_data = team_stats[(team_stats['team'] == team_name) & (team_stats['year'] == year)]
        
        if not team_data.empty:
            # Check if data is in long format (with statName column) or wide format
            if 'statName' in team_data.columns and 'stat' in team_data.columns:
                # Long format - pivot the data
                stats_dict = team_data.set_index('statName')['stat'].to_dict()
                
                # Extract key stats with default values
                features.update({
                    'points_per_game': stats_dict.get('pointsPerGame', 0),
                    'total_yards': stats_dict.get('totalYards', 0),
                    'passing_yards': stats_dict.get('passingYards', 0),
                    'rushing_yards': stats_dict.get('rushingYards', 0),
                    'turnovers': stats_dict.get('turnoversLost', 0),
                    'third_down_conv': stats_dict.get('thirdDownConversions', 0),
                    'penalties': stats_dict.get('penalties', 0),
                })
            else:
                # Wide format - use column names directly
                row = team_data.iloc[0]
                features.update({
                    'points_per_game': row.get('pointsPerGame', 0),
                    'total_yards': row.get('totalYards', 0),
                    'passing_yards': row.get('passingYards', 0),
                    'rushing_yards': row.get('rushingYards', 0),
                    'turnovers': row.get('turnoversLost', 0),
                })
    
    # Advanced stats
    adv_stats = cfb_data['advanced_stats']
    if not adv_stats.empty:
        adv_data = adv_stats[(adv_stats['team'] == team_name) & (adv_stats['year'] == year)]
        if not adv_data.empty:
            row = adv_data.iloc[0]
            features.update({
                'off_success_rate': row.get('offense.successRate', 0),
                'off_explosiveness': row.get('offense.explosiveness', 0),
                'def_success_rate': row.get('defense.successRate', 0),
                'def_explosiveness': row.get('defense.explosiveness', 0),
            })
    
    # Ratings
    for rating_type in ['sp', 'elo', 'fpi']:
        ratings = cfb_data.get(f'ratings_{rating_type}')
        if ratings is not None and not ratings.empty:
            rating_data = ratings[(ratings['team'] == team_name) & (ratings['year'] == year)]
            if not rating_data.empty:
                row = rating_data.iloc[0]
                if rating_type == 'sp':
                    features[f'{rating_type}_rating'] = row.get('rating', 0)
                    features[f'{rating_type}_offense'] = row.get('offense.rating', 0)
                    features[f'{rating_type}_defense'] = row.get('defense.rating', 0)
                elif rating_type == 'elo':
                    features[f'{rating_type}_rating'] = row.get('elo', 1500)  # Default Elo
                elif rating_type == 'fpi':
                    features[f'{rating_type}_rating'] = row.get('fpi', 0)
    
    # Fill any missing values with reasonable defaults
    for key, value in features.items():
        if pd.isna(value) or value is None:
            if 'rating' in key.lower():
                features[key] = 0 if 'elo' not in key else 1500
            else:
                features[key] = 0
    
    return features

def create_training_features(cfb_data):
    """Create features using historical data for training"""
    print("Creating training features from historical data...")
    
    # We'll create a synthetic training dataset using team stats
    # This approach uses team strength differences to predict outcomes
    
    training_features = []
    teams_by_year = {}
    
    # Get all teams by year from team stats
    if not cfb_data['team_stats'].empty:
        teams_stats = cfb_data['team_stats']
        for year in teams_stats['year'].unique():
            year_teams = teams_stats[teams_stats['year'] == year]['team'].unique()
            teams_by_year[year] = list(year_teams)
    
    print(f"Found teams for years: {list(teams_by_year.keys())}")
    
    # Create synthetic matchups for training
    import random
    random.seed(42)  # For reproducible results
    
    for year, teams in teams_by_year.items():
        if len(teams) < 2:
            continue
            
        print(f"Creating training data for {year} with {len(teams)} teams...")
        
        # Create a reasonable number of matchups per year
        num_matchups = min(500, len(teams) * 2)  # Reasonable number
        
        for _ in range(num_matchups):
            # Randomly select two different teams
            home_team, away_team = random.sample(teams, 2)
            
            # Get team features
            home_features = create_team_features(home_team, year, cfb_data)
            away_features = create_team_features(away_team, year, cfb_data)
            
            # Skip if no features available
            if not home_features or not away_features:
                continue
            
            # Create synthetic outcome based on team strength
            # This is a simplified model - better teams should win more often
            home_strength = (
                home_features.get('sp_rating', 0) + 
                home_features.get('elo_rating', 1500) / 1500 * 100 +
                home_features.get('points_per_game', 20)
            )
            
            away_strength = (
                away_features.get('sp_rating', 0) + 
                away_features.get('elo_rating', 1500) / 1500 * 100 +
                away_features.get('points_per_game', 20)
            )
            
            # Add home field advantage
            home_strength += 3  # 3-point home field advantage
            
            # Determine winner with some randomness
            strength_diff = home_strength - away_strength
            win_probability = 1 / (1 + np.exp(-strength_diff / 10))  # Sigmoid function
            home_wins = 1 if random.random() < win_probability else 0
            
            # Create feature row
            game_features = {
                'year': year,
                'week': random.randint(1, 12),
                'neutral_site': 0,  # Assume home games for training
                'home_win': home_wins,
            }
            
            # Add team features
            for key, value in home_features.items():
                game_features[f'home_{key}'] = value
            
            for key, value in away_features.items():
                game_features[f'away_{key}'] = value
            
            # Add differential features
            for key in home_features.keys():
                if key in away_features:
                    game_features[f'diff_{key}'] = home_features[key] - away_features[key]
            
            training_features.append(game_features)
    
    if not training_features:
        print("❌ No training features could be created!")
        return pd.DataFrame()
    
    df = pd.DataFrame(training_features)
    print(f"✅ Created {len(df)} training examples")
    print(f"Home win rate: {df['home_win'].mean():.3f}")
    
    return df

def prepare_prediction_features(cfb_data):
    """Prepare features for 2025 predictions using the games data"""
    print("Preparing features for prediction...")
    
    games = cfb_data['games']
    if games.empty:
        print("No games data available for prediction!")
        return pd.DataFrame()
    
    prediction_features = []
    
    for idx, game in games.iterrows():
        # Get team names
        home_team = game.get('home_team') or game.get('home') or game.get('team_home')
        away_team = game.get('away_team') or game.get('away') or game.get('team_away')
        
        if not home_team or not away_team:
            continue
        
        # Use most recent year for team features (likely 2024)
        feature_year = 2024  # Use most recent complete season
        
        # Get team features
        home_features = create_team_features(home_team, feature_year, cfb_data)
        away_features = create_team_features(away_team, feature_year, cfb_data)
        
        if not home_features or not away_features:
            continue
        
        # Create game features
        game_features = {
            'year': game.get('year', 2025),
            'week': game.get('week', 1),
            'neutral_site': 1 if game.get('neutral_site', False) else 0,
            'home_team_name': home_team,
            'away_team_name': away_team,
        }
        
        # Add team features
        for key, value in home_features.items():
            game_features[f'home_{key}'] = value
        
        for key, value in away_features.items():
            game_features[f'away_{key}'] = value
        
        # Add differential features
        for key in home_features.keys():
            if key in away_features:
                game_features[f'diff_{key}'] = home_features[key] - away_features[key]
        
        prediction_features.append(game_features)
    
    if not prediction_features:
        print("❌ No prediction features could be created!")
        return pd.DataFrame()
    
    df = pd.DataFrame(prediction_features)
    print(f"✅ Created features for {len(df)} games to predict")
    
    return df

# Create both training and prediction features
print("Creating training dataset from historical team performance...")
training_df = create_training_features(cfb_data)

print("\nCreating prediction dataset from scheduled games...")
prediction_df = prepare_prediction_features(cfb_data)

if not training_df.empty:
    print(f"\nTraining data shape: {training_df.shape}")
    print(f"Training features: {len([c for c in training_df.columns if c.startswith(('home_', 'away_', 'diff_'))])} features")

if not prediction_df.empty:
    print(f"Prediction data shape: {prediction_df.shape}")
    print(f"Games to predict: {len(prediction_df)}")
    print(f"Sample matchups:")
    for i, row in prediction_df.head(5).iterrows():
        print(f"  {row.get('home_team_name', 'TBD')} vs {row.get('away_team_name', 'TBD')}")
else:
    print("❌ No prediction data available")
```

```python
# Cell 8: Model Training
def train_models_new_approach(X_train, y_train, feature_columns):
    """Train models using the synthetic training data"""
    
    if X_train is None or y_train is None:
        print("❌ No training data available!")
        return None, None, None, None
    
    print(f"Training models with {len(X_train)} examples and {len(feature_columns)} features...")
    
    # Split for validation
    X_train_split, X_val_split, y_train_split, y_val_split = train_test_split(
        X_train, y_train, test_size=0.2, random_state=42, stratify=y_train
    )
    
    # Scale features
    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train_split)
    X_val_scaled = scaler.transform(X_val_split)
    
    # Models to try
    models = {
        'Random Forest': RandomForestClassifier(n_estimators=100, max_depth=10, random_state=42),
        'Gradient Boosting': GradientBoostingClassifier(n_estimators=100, max_depth=6, random_state=42),
        'Logistic Regression': LogisticRegression(random_state=42, max_iter=1000)
    }
    
    results = {}
    
    for name, model in models.items():
        print(f"\nTraining {name}...")
        
        # Train model
        if name == 'Logistic Regression':
            model.fit(X_train_scaled, y_train_split)
            y_pred = model.predict(X_val_scaled)
            y_pred_proba = model.predict_proba(X_val_scaled)[:, 1]
        else:
            model.fit(X_train_split, y_train_split)
            y_pred = model.predict(X_val_split)
            y_pred_proba = model.predict_proba(X_val_split)[:, 1]
        
        # Evaluate
        accuracy = accuracy_score(y_val_split, y_pred)
        auc_score = roc_auc_score(y_val_split, y_pred_proba)
        
        results[name] = {
            'model': model,
            'accuracy': accuracy,
            'auc': auc_score,
            'predictions': y_pred,
            'probabilities': y_pred_proba
        }
        
        print(f"  Accuracy: {accuracy:.3f}")
        print(f"  AUC: {auc_score:.3f}")
    
    # Select best model
    best_model_name = max(results.keys(), key=lambda k: results[k]['auc'])
    best_model = results[best_model_name]['model']
    
    print(f"\n✅ Best Model: {best_model_name}")
    print(f"Best AUC: {results[best_model_name]['auc']:.3f}")
    
    # Feature importance for tree-based models
    if hasattr(best_model, 'feature_importances_'):
        feature_importance = pd.DataFrame({
            'feature': feature_columns,
            'importance': best_model.feature_importances_
        }).sort_values('importance', ascending=False)
        
        print(f"\nTop 10 Most Important Features:")
        print(feature_importance.head(10).to_string(index=False))
        
        # Plot feature importance
        plt.figure(figsize=(12, 8))
        top_features = feature_importance.head(15)
        plt.barh(range(len(top_features)), top_features['importance'])
        plt.yticks(range(len(top_features)), top_features['feature'])
        plt.xlabel('Feature Importance')
        plt.title(f'Top 15 Feature Importances - {best_model_name}')
        plt.gca().invert_yaxis()
        plt.tight_layout()
        plt.show()
    
    return results, best_model, best_model_name, scaler

# Train models if we have training data
if X_train is not None and y_train is not None:
    model_results, best_model, best_model_name, scaler = train_models_new_approach(X_train, y_train, feature_columns)
    
    if model_results:
        print("\n" + "="*50)
        print("MODEL TRAINING COMPLETE!")
        print("="*50)
        
        # Model comparison plot
        fig, axes = plt.subplots(1, 2, figsize=(15, 5))
        
        model_names = list(model_results.keys())
        accuracies = [model_results[name]['accuracy'] for name in model_names]
        aucs = [model_results[name]['auc'] for name in model_names]
        
        # Accuracy comparison
        axes[0].bar(model_names, accuracies, color=['skyblue', 'lightgreen', 'salmon'])
        axes[0].set_title('Model Accuracy Comparison')
        axes[0].set_ylabel('Accuracy')
        axes[0].set_ylim(0.5, 1.0)
        axes[0].tick_params(axis='x', rotation=45)
        
        # AUC comparison
        axes[1].bar(model_names, aucs, color=['skyblue', 'lightgreen', 'salmon'])
        axes[1].set_title('Model AUC Comparison')
        axes[1].set_ylabel('AUC Score')
        axes[1].set_ylim(0.5, 1.0)
        axes[1].tick_params(axis='x', rotation=45)
        
        plt.tight_layout()
        plt.show()
    else:
        print("❌ Model training failed!")
else:
    print("❌ Cannot train models - no training data available")
    model_results, best_model, best_model_name, scaler = None, None, None, None
```

```python
# Cell 9: Feature Analysis and Correlation (Fixed)
def analyze_features(X, y, feature_columns):
    """Analyze feature importance and correlations"""
    
    if X is None or y is None:
        print("❌ No training data available for analysis")
        return pd.DataFrame()
    
    # Feature correlation with target
    correlations = []
    for col in feature_columns:
        corr = X[col].corr(y)
        correlations.append({'feature': col, 'correlation': abs(corr)})
    
    corr_df = pd.DataFrame(correlations).sort_values('correlation', ascending=False)
    
    # Plot top correlations
    plt.figure(figsize=(12, 8))
    top_features = corr_df.head(20)
    plt.barh(range(len(top_features)), top_features['correlation'])
    plt.yticks(range(len(top_features)), top_features['feature'])
    plt.xlabel('Absolute Correlation with Home Win')
    plt.title('Top 20 Features by Correlation with Home Win')
    plt.gca().invert_yaxis()
    plt.tight_layout()
    plt.show()
    
    # Feature correlation matrix (top features)
    top_feature_names = top_features['feature'].head(15).tolist()
    corr_matrix = X[top_feature_names].corr()
    
    plt.figure(figsize=(12, 10))
    sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0, fmt='.2f')
    plt.title('Feature Correlation Matrix (Top 15 Features)')
    plt.tight_layout()
    plt.show()
    
    return corr_df

# Use the correct variable names from Cell 7
if X_train is not None and y_train is not None:
    print("Analyzing feature correlations...")
    feature_correlations = analyze_features(X_train, y_train, feature_columns)
    
    if not feature_correlations.empty:
        print(f"\nTop 10 Features by Correlation:")
        print(feature_correlations.head(10).to_string(index=False))
    else:
        print("❌ Could not analyze feature correlations")
else:
    print("❌ No training data available for feature analysis")
    feature_correlations = pd.DataFrame()
```

```python
# Cell 10: Generate Predictions for 2025 Games
def make_2025_predictions(prediction_df, best_model, best_model_name, scaler, feature_columns):
    """Generate predictions for 2025 games"""
    
    if prediction_df.empty or best_model is None:
        print("❌ Cannot make predictions - missing data or model")
        return pd.DataFrame()
    
    print(f"Making predictions for {len(prediction_df)} games using {best_model_name}...")
    
    predictions = []
    
    for idx, game in prediction_df.iterrows():
        # Extract features for this game
        game_features = []
        for col in feature_columns:
            game_features.append(game.get(col, 0))
        
        # Make prediction
        X_pred = np.array(game_features).reshape(1, -1)
        
        # Scale if using logistic regression
        if best_model_name == 'Logistic Regression':
            X_pred = scaler.transform(X_pred)
        
        # Get prediction and probabilities
        prediction = best_model.predict(X_pred)[0]
        probabilities = best_model.predict_proba(X_pred)[0]
        
        # Create prediction record
        pred_record = {
            'Home Team': game.get('home_team_name', 'TBD'),
            'Visiting Team': game.get('away_team_name', 'TBD'),
            'Week': game.get('week', 1),
            'Predicted Winner': game.get('home_team_name', 'TBD') if prediction == 1 else game.get('away_team_name', 'TBD'),
            'Home Win Probability': probabilities[1],
            'Away Win Probability': probabilities[0],
            'Confidence': max(probabilities),
            'Predicted Home Win': int(prediction)
        }
        
        predictions.append(pred_record)
    
    predictions_df = pd.DataFrame(predictions)
    print(f"✅ Generated {len(predictions_df)} predictions")
    
    return predictions_df

# Generate predictions if we have everything we need
if 'prediction_df' in locals() and not prediction_df.empty and best_model is not None:
    predictions_2025 = make_2025_predictions(prediction_df, best_model, best_model_name, scaler, feature_columns)
    
    if not predictions_2025.empty:
        print("\n" + "="*60)
        print("2025 PREDICTIONS GENERATED!")
        print("="*60)
        
        # Save predictions
        predictions_2025.to_csv(f'{DATA_DIR}/2025_predictions.csv', index=False)
        print(f"Predictions saved to {DATA_DIR}/2025_predictions.csv")
        
        # Display summary statistics
        print(f"\nPREDICTION SUMMARY:")
        print(f"Total games predicted: {len(predictions_2025)}")
        print(f"Predicted home wins: {predictions_2025['Predicted Home Win'].sum()}")
        print(f"Predicted away wins: {len(predictions_2025) - predictions_2025['Predicted Home Win'].sum()}")
        print(f"Average confidence: {predictions_2025['Confidence'].mean():.3f}")
        
        # High confidence predictions
        high_conf = predictions_2025[predictions_2025['Confidence'] > 0.7].sort_values('Confidence', ascending=False)
        print(f"\nHIGH CONFIDENCE PREDICTIONS (>70%):")
        print("-" * 60)
        for _, game in high_conf.head(10).iterrows():
            loser = game['Home Team'] if game['Predicted Winner'] != game['Home Team'] else game['Visiting Team']
            print(f"{game['Predicted Winner']:20s} beats {loser:20s} ({game['Confidence']:.1%})")
        
        # Upset predictions (away team heavily favored)
        upsets = predictions_2025[predictions_2025['Home Win Probability'] < 0.3].sort_values('Home Win Probability')
        if not upsets.empty:
            print(f"\nUPSET ALERTS (Away team >70% chance):")
            print("-" * 60)
            for _, game in upsets.head(5).iterrows():
                print(f"{game['Visiting Team']:20s} @ {game['Home Team']:20s} ({game['Away Win Probability']:.1%} away win)")
        
        # Visualization
        fig, axes = plt.subplots(2, 2, figsize=(15, 12))
        
        # Home win probability distribution
        axes[0,0].hist(predictions_2025['Home Win Probability'], bins=20, alpha=0.7, edgecolor='black')
        axes[0,0].set_xlabel('Home Win Probability')
        axes[0,0].set_ylabel('Number of Games')
        axes[0,0].set_title('Distribution of Home Win Probabilities')
        axes[0,0].axvline(x=0.5, color='red', linestyle='--', label='Even odds')
        axes[0,0].legend()
        
        # Confidence distribution
        axes[0,1].hist(predictions_2025['Confidence'], bins=20, alpha=0.7, edgecolor='black', color='green')
        axes[0,1].set_xlabel('Prediction Confidence')
        axes[0,1].set_ylabel('Number of Games')
        axes[0,1].set_title('Distribution of Prediction Confidence')
        
        # Home vs Away predictions
        home_wins = predictions_2025['Predicted Home Win'].sum()
        away_wins = len(predictions_2025) - home_wins
        axes[1,0].pie([home_wins, away_wins], labels=['Home Wins', 'Away Wins'], autopct='%1.1f%%')
        axes[1,0].set_title('Predicted Home vs Away Wins')
        
        # Confidence by week
        if 'Week' in predictions_2025.columns:
            conf_by_week = predictions_2025.groupby('Week')['Confidence'].mean()
            axes[1,1].bar(conf_by_week.index, conf_by_week.values)
            axes[1,1].set_xlabel('Week')
            axes[1,1].set_ylabel('Average Confidence')
            axes[1,1].set_title('Average Prediction Confidence by Week')
        else:
            axes[1,1].text(0.5, 0.5, 'Week data not available', ha='center', va='center', transform=axes[1,1].transAxes)
        
        plt.tight_layout()
        plt.show()
        
    else:
        print("❌ Failed to generate predictions")
elif best_model is None:
    print("❌ Cannot generate predictions - no trained model available")
elif 'prediction_df' not in locals() or prediction_df.empty:
    print("❌ Cannot generate predictions - no prediction data available")
else:
    print("❌ Cannot generate predictions - missing required data")
    predictions_2025 = pd.DataFrame()

print(f"\nSTATUS UPDATE:")
print(f"Model trained: {'✅' if 'best_model' in locals() and best_model is not None else '❌'}")
print(f"Predictions generated: {'✅' if 'predictions_2025' in locals() and not predictions_2025.empty else '❌'}")
```

```python
# Cell 11: Custom Prediction Function
def predict_custom_matchup(team1, team2, week=1, neutral_site=False, year=2024):
    """Predict outcome for a custom matchup"""
    
    if 'best_model' not in locals() or best_model is None:
        print("❌ No trained model available for predictions")
        return None
    
    print(f"\nPREDICTING: {team1} vs {team2}")
    print(f"Week: {week}, Neutral Site: {'Yes' if neutral_site else 'No'}")
    
    # Get team features (using most recent year)
    home_features = create_team_features(team1, year, cfb_data)
    away_features = create_team_features(team2, year, cfb_data)
    
    if not home_features or not away_features:
        print(f"❌ Could not find team data for {team1} or {team2}")
        return None
    
    # Create game features
    game_features_dict = {
        'week': week,
        'neutral_site': 1 if neutral_site else 0,
    }
    
    # Add team features
    for key, value in home_features.items():
        game_features_dict[f'home_{key}'] = value
    
    for key, value in away_features.items():
        game_features_dict[f'away_{key}'] = value
    
    # Add differential features
    for key in home_features.keys():
        if key in away_features:
            game_features_dict[f'diff_{key}'] = home_features[key] - away_features[key]
    
    # Create feature vector
    game_features = []
    for col in feature_columns:
        game_features.append(game_features_dict.get(col, 0))
    
    # Make prediction
    X_pred = np.array(game_features).reshape(1, -1)
    
    if best_model_name == 'Logistic Regression':
        X_pred = scaler.transform(X_pred)
    
    prediction = best_model.predict(X_pred)[0]
    probabilities = best_model.predict_proba(X_pred)[0]
    
    winner = team1 if prediction == 1 else team2
    win_prob = probabilities[1] if prediction == 1 else probabilities[0]
    
    print(f"🏆 Predicted Winner: {winner}")
    print(f"📊 Win Probability: {win_prob:.1%}")
    print(f"🎯 Confidence: {max(probabilities):.1%}")
    print(f"📈 {team1} win probability: {probabilities[1]:.1%}")
    print(f"📈 {team2} win probability: {probabilities[0]:.1%}")
    
    return {
        'winner': winner,
        'home_win_prob': probabilities[1],
        'away_win_prob': probabilities[0],
        'confidence': max(probabilities)
    }

# Example predictions if model is available
if 'best_model' in locals() and best_model is not None:
    print("\n" + "="*60)
    print("CUSTOM PREDICTION EXAMPLES")
    print("="*60)
    
    # Try some big matchups
    example_matchups = [
        ('Alabama', 'Georgia'),
        ('Ohio State', 'Michigan'),
        ('Texas', 'Oklahoma'),
        ('USC', 'UCLA'),
        ('Notre Dame', 'Stanford')
    ]
    
    for home, away in example_matchups:
        try:
            predict_custom_matchup(home, away, week=1)
        except Exception as e:
            print(f"❌ Error predicting {home} vs {away}: {e}")
else:
    print("❌ No trained model available for custom predictions")

print("\n" + "="*80)
print("🏈 COLLEGE FOOTBALL PREDICTION SYSTEM STATUS 🏈")
print("="*80)

# Final status check
if 'best_model' in locals() and best_model is not None:
    print("✅ Model trained successfully")
    print(f"✅ Best model: {best_model_name}")
    if 'predictions_2025' in locals() and not predictions_2025.empty:
        print(f"✅ Generated {len(predictions_2025)} predictions for 2025")
        print(f"✅ Results saved to {DATA_DIR}/2025_predictions.csv")
    else:
        print("❌ No 2025 predictions generated")
    print("\n🎯 You can now use predict_custom_matchup() to predict any game!")
    print("   Example: predict_custom_matchup('Alabama', 'Georgia', week=1)")
else:
    print("❌ Model training failed - check your data files")

print(f"\n📁 Check the '{DATA_DIR}' directory for output files")
print("🏆 Happy predicting!")
```

```python
# Cell 12: 2025 Prediction Setup
def load_2025_schedule(data_dir):
    """Load 2025 schedule or create sample matchups"""
    try:
        games_2025 = pd.read_csv(f'{data_dir}/games_2025.csv')
        print(f"Loaded {len(games_2025)} games for 2025 season")
    except FileNotFoundError:
        print("2025 schedule not found. Creating sample matchups...")
        
        # Get teams from recent data
        try:
            teams_df = pd.read_csv(f'{data_dir}/teams_fbs.csv')
            sample_teams = teams_df['school'].head(20).tolist()
        except:
            # Fallback to common teams
            sample_teams = [
                'Alabama', 'Georgia', 'Ohio State', 'Michigan', 'Texas', 'USC', 
                'Oklahoma', 'Notre Dame', 'Penn State', 'Florida', 'LSU', 
                'Auburn', 'Tennessee', 'Wisconsin', 'Oregon', 'Washington',
                'Miami', 'Virginia Tech', 'North Carolina', 'Clemson'
            ]
        
        # Create sample matchups
        games_2025 = []
        for i in range(0, len(sample_teams), 2):
            if i + 1 < len(sample_teams):
                games_2025.append({
                    'home_team': sample_teams[i],
                    'away_team': sample_teams[i + 1],
                    'week': 1,
                    'season_type': 'regular'
                })
        
        # Add some more interesting matchups
        marquee_matchups = [
            ('Alabama', 'Georgia'), ('Ohio State', 'Michigan'),
            ('Texas', 'Oklahoma'), ('USC', 'Notre Dame'),
            ('Penn State', 'Wisconsin'), ('Florida', 'LSU')
        ]
        
        for home, away in marquee_matchups:
            games_2025.append({
                'home_team': home,
                'away_team': away,
                'week': 2,
                'season_type': 'regular'
            })
        
        games_2025 = pd.DataFrame(games_2025)
        print(f"Created {len(games_2025)} sample matchups")
    
    return games_2025

games_2025 = load_2025_schedule(DATA_DIR)
print("Sample 2025 matchups:")
print(games_2025.head(10))
```

```python
# Cell 16: Final Summary and Custom Prediction Function
def predict_custom_game(team1, team2, week=1, neutral_site=False):
    """Predict outcome for a custom matchup"""
    
    if 'best_model' not in locals() or best_model is None:
        print("❌ No trained model available for predictions")
        return None
    
    print(f"\nCUSTOM PREDICTION: {team1} vs {team2}")
    print(f"Week: {week}, Neutral Site: {'Yes' if neutral_site else 'No'}")
    
    # Get team features (using most recent year available, likely 2024)
    home_features = create_team_features(team1, 2024, cfb_data)
    away_features = create_team_features(team2, 2024, cfb_data)
    
    if not home_features or not away_features:
        print(f"❌ Could not find team data for {team1} or {team2}")
        print("Available teams might be limited to those in your dataset")
        return None
    
    # Create game features
    game_features_dict = {
        'week': week,
        'neutral_site': 1 if neutral_site else 0,
    }
    
    # Add team features
    for key, value in home_features.items():
        game_features_dict[f'home_{key}'] = value
    
    for key, value in away_features.items():
        game_features_dict[f'away_{key}'] = value
    
    # Add differential features
    for key in home_features.keys():
        if key in away_features:
            game_features_dict[f'diff_{key}'] = home_features[key] - away_features[key]
    
    # Create feature vector
    game_features = []
    for col in feature_columns:
        game_features.append(game_features_dict.get(col, 0))
    
    # Make prediction
    X_pred = np.array(game_features).reshape(1, -1)
    
    if best_model_name == 'Logistic Regression':
        X_pred = scaler.transform(X_pred)
    
    prediction = best_model.predict(X_pred)[0]
    probabilities = best_model.predict_proba(X_pred)[0]
    
    winner = team1 if prediction == 1 else team2
    win_prob = probabilities[1] if prediction == 1 else probabilities[0]
    
    print(f"🏆 Predicted Winner: {winner}")
    print(f"📊 Win Probability: {win_prob:.1%}")
    print(f"🎯 Confidence: {max(probabilities):.1%}")
    print(f"📈 {team1} (Home) win probability: {probabilities[1]:.1%}")
    print(f"📈 {team2} (Away) win probability: {probabilities[0]:.1%}")
    
    return {
        'winner': winner,
        'home_win_prob': probabilities[1],
        'away_win_prob': probabilities[0],
        'confidence': max(probabilities)
    }

# Test the custom prediction function with some examples
if 'best_model' in locals() and best_model is not None:
    print("=" * 70)
    print("🏈 TESTING CUSTOM PREDICTION FUNCTION 🏈")
    print("=" * 70)
    
    # Test matchups
    test_matchups = [
        ('Alabama', 'Georgia'),
        ('Ohio State', 'Michigan'), 
        ('Texas', 'Oklahoma')
    ]
    
    for home, away in test_matchups:
        try:
            predict_custom_game(home, away, week=1)
        except Exception as e:
            print(f"❌ Error predicting {home} vs {away}: {e}")
    
    print("\n" + "=" * 70)
    print("🎯 PREDICTION SYSTEM READY!")
    print("=" * 70)
    print("✅ You can now use predict_custom_game() to predict any matchup!")
    print("   Example: predict_custom_game('Alabama', 'Georgia', week=1)")
    print("   Example: predict_custom_game('USC', 'UCLA', week=12, neutral_site=True)")
    
else:
    print("❌ Custom prediction function not available - no trained model")

# Final system status
print(f"\n📊 FINAL SYSTEM STATUS:")
print(f"{'='*50}")

status_items = [
    ("Data loaded", 'cfb_data' in locals() and cfb_data),
    ("Training data created", 'training_df' in locals() and not training_df.empty),
    ("Model trained", 'best_model' in locals() and best_model is not None),
    ("2025 predictions generated", 'predictions_2025' in locals() and not predictions_2025.empty),
    ("Results saved", True)  # Assume true if we got this far
]

for item, status in status_items:
    print(f"  {item}: {'✅' if status else '❌'}")

if 'best_model_name' in locals():
    print(f"\n🏆 Best Model: {best_model_name}")

if 'predictions_2025' in locals() and not predictions_2025.empty:
    print(f"📈 Total 2025 Predictions: {len(predictions_2025)}")

print(f"\n📁 Output files saved to: {DATA_DIR}/")
print("🏈 College Football Prediction System Complete!")
```

```python
def get_team_conference(team_name):
    """Determine which conference a team belongs to"""
    
    conferences = {
        'SEC': ['Alabama', 'Arkansas', 'Auburn', 'Florida', 'Georgia', 'Kentucky', 
                'LSU', 'Mississippi', 'Mississippi State', 'Missouri', 'South Carolina', 
                'Tennessee', 'Texas A&M', 'Vanderbilt', 'Texas', 'Oklahoma'],
        
        'Big Ten': ['Illinois', 'Indiana', 'Iowa', 'Maryland', 'Michigan', 'Michigan State',
                   'Minnesota', 'Nebraska', 'Northwestern', 'Ohio State', 'Penn State', 
                   'Purdue', 'Rutgers', 'Wisconsin', 'Oregon', 'UCLA', 'USC', 'Washington'],
        
        'ACC': ['Boston College', 'Clemson', 'Duke', 'Florida State', 'Georgia Tech',
                'Louisville', 'Miami', 'North Carolina', 'NC State', 'Pittsburgh', 
                'Syracuse', 'Virginia', 'Virginia Tech', 'Wake Forest', 'Notre Dame', 
                'Stanford', 'California', 'SMU'],
        
        'Big 12': ['Arizona', 'Arizona State', 'Baylor', 'BYU', 'Cincinnati', 'Colorado',
                  'Houston', 'Iowa State', 'Kansas', 'Kansas State', 'Oklahoma State', 
                  'TCU', 'Texas Tech', 'UCF', 'Utah', 'West Virginia'],
        
        'Pac-12': ['Oregon State', 'Washington State'],  # Remaining Pac-12
        
        'AAC': ['East Carolina', 'Memphis', 'Navy', 'SMU', 'South Florida', 'Temple', 
                'Tulane', 'Tulsa', 'UAB', 'UTSA'],
        
        'Mountain West': ['Air Force', 'Boise State', 'Colorado State', 'Fresno State', 
                         'Hawaii', 'Nevada', 'New Mexico', 'San Diego State', 'San Jose State', 
                         'UNLV', 'Utah State', 'Wyoming'],
        
        'MAC': ['Akron', 'Ball State', 'Bowling Green', 'Buffalo', 'Central Michigan',
                'Eastern Michigan', 'Kent State', 'Miami (OH)', 'Northern Illinois', 
                'Ohio', 'Toledo', 'Western Michigan'],
        
        'Sun Belt': ['Appalachian State', 'Arkansas State', 'Coastal Carolina', 'Georgia Southern',
                    'Georgia State', 'James Madison', 'Louisiana', 'Louisiana Monroe', 
                    'Marshall', 'Old Dominion', 'South Alabama', 'Southern Miss', 'Texas State', 'Troy'],
        
        'CUSA': ['Charlotte', 'FAU', 'FIU', 'Louisiana Tech', 'Middle Tennessee', 'North Texas',
                'Rice', 'UAB', 'UTEP', 'Western Kentucky'],
        
        'Independent': ['Army', 'Liberty', 'New Mexico State', 'UConn', 'UMass']
    }
    
    for conf, teams in conferences.items():
        if team_name in teams:
            return conf, teams
    
    return 'Unknown', []

def generate_realistic_schedule(team_name):
    """Generate a realistic 12-game schedule for any team"""
    
    # Get available teams from your data
    if 'cfb_data' in locals() and not cfb_data['team_stats'].empty:
        all_teams = list(cfb_data['team_stats']['team'].unique())
    else:
        # Fallback list if data not available
        all_teams = ['Alabama', 'Arkansas', 'Auburn', 'Florida', 'Georgia', 'Kentucky', 
                'LSU', 'Mississippi', 'Mississippi State', 'Missouri', 'South Carolina', 
                'Tennessee', 'Texas A&M', 'Vanderbilt', 'Texas', 'Oklahoma', 'Illinois', 'Indiana', 'Iowa', 'Maryland', 'Michigan', 'Michigan State',
                   'Minnesota', 'Nebraska', 'Northwestern', 'Ohio State', 'Penn State', 
                   'Purdue', 'Rutgers', 'Wisconsin', 'Oregon', 'UCLA', 'USC', 'Washington', 'Boston College', 'Clemson', 'Duke', 'Florida State', 'Georgia Tech',
                'Louisville', 'Miami', 'North Carolina', 'NC State', 'Pittsburgh', 
                'Syracuse', 'Virginia', 'Virginia Tech', 'Wake Forest', 'Notre Dame', 
                'Stanford', 'California', 'SMU', 'Arizona', 'Arizona State', 'Baylor', 'BYU', 'Cincinnati', 'Colorado',
                  'Houston', 'Iowa State', 'Kansas', 'Kansas State', 'Oklahoma State', 
                  'TCU', 'Texas Tech', 'UCF', 'Utah', 'West Virginia', 'Oregon State', 'Washington State', 'East Carolina', 'Memphis', 'Navy', 'SMU', 'South Florida', 'Temple', 
                'Tulane', 'Tulsa', 'UAB', 'UTSA', 'Air Force', 'Boise State', 'Colorado State', 'Fresno State', 
                         'Hawaii', 'Nevada', 'New Mexico', 'San Diego State', 'San Jose State', 
                         'UNLV', 'Utah State', 'Wyoming', 'Akron', 'Ball State', 'Bowling Green', 'Buffalo', 'Central Michigan',
                'Eastern Michigan', 'Kent State', 'Miami (OH)', 'Northern Illinois', 
                'Ohio', 'Toledo', 'Western Michigan', 'Appalachian State', 'Arkansas State', 'Coastal Carolina', 'Georgia Southern',
                    'Georgia State', 'James Madison', 'Louisiana', 'Louisiana Monroe', 
                    'Marshall', 'Old Dominion', 'South Alabama', 'Southern Miss', 'Texas State', 'Troy', 'Charlotte', 'FAU', 'FIU', 'Louisiana Tech', 'Middle Tennessee', 'North Texas',
                'Rice', 'UAB', 'UTEP', 'Western Kentucky', 'Army', 'Liberty', 'New Mexico State', 'UConn', 'UMass']
    
    if team_name not in all_teams:
        print(f"❌ {team_name} not found in dataset")
        print(f"Available teams: {all_teams[:10]}... and {len(all_teams)-10} more")
        return None
    
    conference, conf_teams = get_team_conference(team_name)
    
    schedule = []
    used_opponents = set()
    
    # Week 1: Usually an easy opponent
    cupcake_opponents = ['Western Kentucky', 'Eastern Kentucky', 'Austin Peay', 'UT Martin',
                        'Jacksonville State', 'Murray State', 'Tennessee State', 'Alabama State',
                        'Akron', 'Bowling Green', 'Ball State', 'Eastern Michigan', 'UMass']
    
    week1_opponent = None
    for opponent in cupcake_opponents:
        if opponent in all_teams and opponent != team_name:
            week1_opponent = opponent
            break
    
    if not week1_opponent:
        # Pick any non-conference opponent
        non_conf = [t for t in all_teams if t not in conf_teams and t != team_name]
        week1_opponent = non_conf[0] if non_conf else 'Generic Opponent'
    
    schedule.append(('Week 1', week1_opponent, True))
    used_opponents.add(week1_opponent)
    
    # Week 2: Another non-conference, slightly harder
    week2_candidates = [t for t in all_teams if t not in conf_teams and t != team_name and t not in used_opponents]
    if week2_candidates:
        week2_opponent = week2_candidates[0]
        schedule.append(('Week 2', week2_opponent, True))
        used_opponents.add(week2_opponent)
    
    # Week 3: Major non-conference game (if power conference)
    if conference in ['SEC', 'Big Ten', 'ACC', 'Big 12']:
        major_opponents = []
        for conf in ['SEC', 'Big Ten', 'ACC', 'Big 12']:
            if conf != conference:
                major_opponents.extend(get_team_conference('Alabama')[1] if conf == 'SEC' else 
                                     get_team_conference('Ohio State')[1] if conf == 'Big Ten' else
                                     get_team_conference('Clemson')[1] if conf == 'ACC' else
                                     get_team_conference('Kansas')[1])
        
        week3_candidates = [t for t in major_opponents if t in all_teams and t != team_name and t not in used_opponents]
        if week3_candidates:
            week3_opponent = week3_candidates[0]
            schedule.append(('Week 3', week3_opponent, False))  # Away game
            used_opponents.add(week3_opponent)
    
    # Weeks 4-12: Conference games (mix of home/away)
    conf_opponents = [t for t in conf_teams if t != team_name and t in all_teams]
    available_conf = [t for t in conf_opponents if t not in used_opponents]
    
    # Fill remaining weeks with conference games
    current_week = len(schedule) + 1
    home_away_pattern = [True, False, True, True, False, True, False, True, False]  # Mix of home/away
    
    for i, opponent in enumerate(available_conf):
        if current_week <= 12:
            is_home = home_away_pattern[i % len(home_away_pattern)]
            schedule.append((f'Week {current_week}', opponent, is_home))
            current_week += 1
    
    # Fill any remaining weeks with additional opponents
    while len(schedule) < 12:
        remaining_opponents = [t for t in all_teams if t != team_name and t not in used_opponents]
        if remaining_opponents:
            opponent = remaining_opponents[0]
            is_home = (current_week % 2 == 1)  # Alternate home/away
            schedule.append((f'Week {current_week}', opponent, is_home))
            used_opponents.add(opponent)
            current_week += 1
        else:
            break
    
    return schedule

def predict_team_season(team_name):
    """
    Predict an entire season for ANY team in your dataset
    Usage: predict_team_season('Alabama') or predict_team_season('Toledo')
    """
    
    print(f"🏈 PREDICTING 2025 SEASON FOR {team_name.upper()}")
    print("=" * 60)
    
    # Generate realistic schedule
    schedule = generate_realistic_schedule(team_name)
    
    if not schedule:
        return None
    
    # Get team's conference info
    conference, conf_teams = get_team_conference(team_name)
    print(f"Conference: {conference}")
    print(f"Generated {len(schedule)}-game schedule")
    print("-" * 60)
    
    # Make predictions for each game
    season_predictions = []
    total_wins = 0
    conf_wins = 0
    total_games = 0
    conf_games = 0
    total_confidence = 0
    
    for week, opponent, is_home in schedule:
        week_num = int(week.split()[1])
        is_conf_game = opponent in conf_teams
        
        try:
            if is_home:
                result = predict_custom_game_fixed(team_name, opponent, week=week_num)
            else:
                result = predict_custom_game_fixed(opponent, team_name, week=week_num)
            
            if result:
                team_won = (result['winner'] == team_name)
                win_prob = result['home_win_prob'] if is_home else result['away_win_prob']
                
                if team_won:
                    total_wins += 1
                    if is_conf_game:
                        conf_wins += 1
                
                total_games += 1
                if is_conf_game:
                    conf_games += 1
                    
                total_confidence += result['confidence']
                
                season_predictions.append({
                    'Week': week,
                    'Opponent': opponent,
                    'Location': 'Home' if is_home else 'Away',
                    'Conference_Game': is_conf_game,
                    'Predicted_Winner': result['winner'],
                    'Team_Won': team_won,
                    'Win_Probability': win_prob,
                    'Confidence': result['confidence']
                })
                
                # Print result with conference indicator
                location = "vs" if is_home else "@"
                status = "W" if team_won else "L"
                conf_indicator = "*" if is_conf_game else " "
                prob_display = f"({win_prob:.1%})"
                print(f"{week}: {status} {location} {opponent:20s} {prob_display} {conf_indicator}")
            
        except Exception as e:
            print(f"❌ Error predicting {week} vs {opponent}: {e}")
    
    # Season summary
    if total_games > 0:
        avg_confidence = total_confidence / total_games
        win_percentage = total_wins / total_games
        conf_win_pct = conf_wins / conf_games if conf_games > 0 else 0
        
        print("\n" + "=" * 60)
        print(f"🏆 {team_name.upper()} SEASON SUMMARY")
        print("=" * 60)
        print(f"Overall Record: {total_wins}-{total_games - total_wins} ({win_percentage:.1%})")
        if conf_games > 0:
            print(f"Conference Record: {conf_wins}-{conf_games - conf_wins} ({conf_win_pct:.1%})")
        print(f"Average Confidence: {avg_confidence:.1%}")
        
        # Bowl eligibility
        if total_wins >= 6:
            print(f"✅ Bowl Eligible!")
        else:
            print(f"❌ Not Bowl Eligible (need {6 - total_wins} more wins)")
        
        # Championship/Playoff implications
        if conference in ['SEC', 'Big Ten', 'ACC', 'Big 12'] and conf_wins >= conf_games * 0.75:
            print(f"🏆 Strong conference championship contender!")
        
        if total_wins >= 11:
            print(f"🏆 College Football Playoff contender!")
        elif total_wins >= 9:
            print(f"🎯 Possible NY6 bowl game")
        
        print(f"\n* = Conference game")
    
    return pd.DataFrame(season_predictions)

def get_available_teams():
    """Get all teams available for prediction"""
    if 'cfb_data' in locals() and not cfb_data['team_stats'].empty:
        teams = sorted(cfb_data['team_stats']['team'].unique())
        return teams
    else:
        return ["Data not loaded - run data loading cells first"]

def predict_random_teams(n=5):
    """Predict seasons for n random teams"""
    import random
    
    all_teams = get_available_teams()
    if len(all_teams) < n:
        n = len(all_teams)
    
    random_teams = random.sample(all_teams, n)
    
    print(f"🎲 PREDICTING SEASONS FOR {n} RANDOM TEAMS")
    print("=" * 70)
    
    results = {}
    for team in random_teams:
        try:
            print(f"\n🏈 {team}")
            print("-" * 40)
            season_df = predict_team_season(team)
            if season_df is not None and not season_df.empty:
                wins = season_df['Team_Won'].sum()
                games = len(season_df)
                results[team] = f"{wins}-{games - wins}"
        except Exception as e:
            print(f"❌ Error with {team}: {e}")
            results[team] = "Error"
    
    return results

# Show available teams
print("🏈 ENHANCED SEASON PREDICTION SYSTEM")
print("=" * 60)
print("✅ Now works with ALL teams in your dataset!")
print("\n💡 USAGE:")
print("predict_team_season('Alabama')      # Major program")  
print("predict_team_season('Toledo')       # Mid-major")
print("predict_team_season('Boise State')  # Group of 5")
print("predict_random_teams(3)             # Try random teams")
print("\nget_available_teams()              # See all available teams")

# Quick check of available teams
teams = get_available_teams()
if isinstance(teams, list) and len(teams) > 1:
    print(f"\n📊 Found {len(teams)} teams in dataset")
    print(f"Sample teams: {teams[:10]}")
    if len(teams) > 10:
        print(f"... and {len(teams) - 10} more")
else:
    print("\n❌ Team data not loaded yet")
```
